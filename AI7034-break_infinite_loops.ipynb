```python
names = ['Max', 'Felix', 'Deniz', 'Selin', 'Lucas', 26, 'Sarah']
 
for name in names:
 
    if type(name) != str:
        print('Found', name)
        break
 
    else:
        print(name, 'is a string!')
```
